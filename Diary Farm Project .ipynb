PART 1: DATA CATEGORIZATION AND CLEANING

```python
#Drop any duplicated data (rows) from the dataset
data_df.drop_duplicates(inplace=True)
```

```python
data_df.shape
```

```python
#Write a script to get numeric columns and non-numeric columns
numeric_cols = data_df.select_dtypes([np.number]).columns
non_numeric_cols = pd.Index(set(data_df.columns).difference(set(data_df.select_dtypes(include=(np.number)).columns)))
print("Numerical columns:", numeric_cols)
print("Non-numerical columns:", non_numeric_cols)
```

```python
#Write a script to find the percentage of the missing values for each column in your dataset
for col in data_df.columns:
    percentage_of_missing = np.mean(data_df[col].isnull())
    print('{} - {}%'.format(col, round(percentage_of_missing*100)))
```

```python
#Impute any numeric feature that has missing values less than 2% by replacing the missing values with the median of the data of the same feature that is not missing
for col in numeric_cols:
    percentage_of_missing = np.mean(data_df[col].isnull())
    if percentage_of_missing < 0.02:
        median = data_df[col].median()
        data_df[col] =  data_df[col].fillna(median)
```

```python
# Impute any numeric feature that has missing values 5% or more by replacing the missing values with the mode of the data of the same feature that is not missing
for col in numeric_cols:
    percentage_of_missing = data_df[col].isnull().mean() * 100
    if percentage_of_missing >= 0.05:
        mode = data_df[col].mode().iloc[0]
        data_df[col] = data_df[col].fillna(mode)
```

```python
# For any non-numeric feature that has less than 5% missing values, replace all missing values with the word '_NA_'.
for col in non_numeric_cols:
    percentage_of_missing = np.mean(data_df[col].isnull())
    if percentage_of_missing < 0.05:
         data_df[col] =  data_df[col].fillna('_NA_')
```

```python
# For any non-numeric feature that has 5% or more missing values, replace all missing values with the most frequent value for this feature
for col in non_numeric_cols:
    percentage_of_missing = np.mean(data_df[col].isnull())
    if percentage_of_missing >= 0.05:
        most_frequent = data_df[col].mode()[0]
        data_df[col] =  data_df[col].fillna(most_frequent)
```

```python
# Replace any remaining missing numeric value with 0 and any remaining missing non-numeric value with the word 'MISSING'.
for col in non_numeric_cols:
    data_df[col] =  data_df[col].fillna('MISSING')
```

```python
for col in numeric_cols:
    data_df[col] =  data_df[col].fillna(0)
```

```python
# For each numeric feature in the dataset, drop observations of this feature that are greater than ( mean+3∗std) or less than ( mean – 3∗std). Where the mean is the mean of the data that is not missing and std is the standard deviation of the datathat is not missing of each feature

for col in numeric_cols:
    mean = data_df[col].mean()
    std = data_df[col].std()
    data_df = data_df[(data_df[col].isnull()) | ((data_df[col] >= mean - 3 * std) & (data_df[col] <= mean + 3 * std))]
```

PART 2: DATA DESCRIPTIVE STATISTICS


```python
#Calculating mean, median, mode, standard deviation, range and IQR for each column
for col in data_df.select_dtypes(include=[np.number]):
    mean = data_df[col].mean()
    median = data_df[col].median()
    mode = data_df[col].mode()[0]
    std = data_df[col].std()
    range = data_df[col].max() - data_df[col].min()
    q1, q3 = np.percentile(data_df[col], [25, 75])
    iqr = q3 - q1
    print(f'\nStatistics for {col}:')
    print(f'Mean: {mean}')
    print(f'Median: {median}')
    print(f'Mode: {mode}')
    print(f'Standard deviation: {std}')
    print(f'Range: {range}')
    print(f'IQR: {iqr}')
```

```python
#For each non-numeric column in your dataset, write the code to perform the most frequent value for column
non_numeric_cols = data_df.select_dtypes(exclude=np.number).columns.tolist()
for col in non_numeric_cols:
    mode = data_df[col].mode()[0]
    unique = data_df[col].unique()
    print(f"Column: {col}")
    print(f"Most frequent value: {mode}")
    print()
```

```python
#For each non-numeric column in your dataset, write the code to perform the unique value for column
non_numeric_cols = data_df.select_dtypes(exclude=np.number).columns.tolist()
for col in non_numeric_cols:
    mode = data_df[col].mode()[0]
    unique = data_df[col].unique()
    print(f"Column: {col}")
    print(f"Unique values: {unique}")
    print()
```

PART 3: DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Top_Customer_Location = data_df['Customer Location'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(Top_Customer_Location.index, Top_Customer_Location.values)
plt.xlabel('Customer Location')
plt.ylabel('Count')
plt.title('Top Customer Locations')
plt.xticks(rotation=45)
plt.show()
```

The bar chart above represent locations where customers resides. And Delhi shows the highest customer locations, followed by Chandigarh and Maharashtra.

```python
sales_channel_counts = data_df['Sales Channel'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sales_channel_counts.values, labels=sales_channel_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Sales Channels')
plt.show()
```

The pie-chart above represent channels where items are sold. Retails shows the highest sales channel followed by the wholesale
i.e Most items sold through retail is 34.2% and wholesale is 34.1%

```python
Top_Product_Sold = data_df['Product Name'].value_counts().head(10)
plt.bar(Top_Product_Sold.index, Top_Product_Sold.values)
plt.xlabel('Product Name')
plt.ylabel('Count')
plt.title('Top Product Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The bar chart above represent Highest Product Sold. Curd shows the highest product purchased followed by Lassi.
